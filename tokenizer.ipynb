```python
corpus = "Nestled in the heart of the bustling city, the quaint caf√© ‚òïÔ∏èüå∏ offers a cozy retreat for weary travelers and locals alike. With its eclectic d√©cor and aromatic brews, it's the perfect spot to unwind and enjoy a moment of tranquility amidst the chaos. Patrons often share stories of adventures, accompanied by a ‚òïÔ∏è in hand and a smile on their faces üòä. The staff, always friendly and attentive, make every visit special, creating a welcoming atmosphere where everyone feels at home. Whether you're seeking a quiet corner to read üìö or a lively place to catch up with friends, this charming caf√© promises a delightful experience every time."
```

```python
tokens = corpus.encode("utf-8")
tokens = list(tokens)
len(corpus), len(tokens)
```

```python
def get_stats(ids):
    count = {}
    for pair in zip(ids, ids[1:]):
        if pair in count:
            count[pair] += 1
        else:
            count[pair] = 1
    return count

stats = get_stats(tokens)

# print(sorted(((v, k) for k, v in stats.items()), reverse=True)[:10])
```

```python
top_pair = max(stats, key=stats.get)
top_pair
```

```python
def merge(ids, top_pair, idx):
    new_tokens = []
    i = 0
    while i < len(ids):
        if i < len(ids) - 1 and ids[i] == top_pair[0] and ids[i + 1] == top_pair[1]:
            new_tokens.append(idx)
            i += 2
        else:
            new_tokens.append(ids[i])
            i += 1
    return new_tokens

print(merge([5, 6, 6, 7, 9, 1], (6, 7), 99))
```

```python
tokens = merge(tokens, top_pair, 256)
print(tokens)
```

```python
len(tokens)
```

```python
vocab_length = 300

num_merges = vocab_length - 255

for i in range(num_merges):
    stats = get_stats(tokens)
    top_pair = max(stats, key=stats.get)

    print(f"Iteration {i + 1}: Merging {top_pair} -> {256 + i}")
    tokens = merge(tokens, top_pair, 256 + i)

tokens
```
